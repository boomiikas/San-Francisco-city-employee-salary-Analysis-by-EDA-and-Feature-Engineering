```python
%matplotlib inline
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
df = pd.read_csv("Salaries.csv")
```

```python
df.drop('Notes', axis=1, inplace=True)
```

```python
df = df.dropna(subset=['BasePay'])                
```

```python
df['Status'] = df['Status'].fillna('Unknown')  
```

```python
df['Benefits'] = pd.to_numeric(df['Benefits'], errors='coerce')
df['Benefits'] = df['Benefits'].fillna(df['Benefits'].median())
```

```python
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce')
df['OtherPay'] = pd.to_numeric(df['OtherPay'], errors='coerce')
```

```python
df = AV.AutoViz("Salaries.csv", depVar="TotalPay", chart_format="svg", max_cols_analyzed=8)
AV.AutoViz("", dfte=df)
```
