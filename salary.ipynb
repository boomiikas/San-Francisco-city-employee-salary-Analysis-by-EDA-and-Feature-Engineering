```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Salaries.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Benefits'].unique()
```

```python
df['Benefits']=pd.to_numeric(df['Benefits'],errors='coerce')
```

```python
df['Benefits'].unique()
```

```python
df['Benefits'] = df['Benefits'].fillna(df['Benefits'].median())
```

```python
df['Benefits'].unique()
```

```python
df['Notes'].unique()
```

```python
df.drop('Notes', axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Status'].unique()
```

```python
df['Status'] = df['Status'].fillna('Unknown')
```

```python
df['Status'].unique()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['BasePay'])
```

```python
df['TotalPay'].hist(bins=50)
plt.xlabel("Total Pay")
plt.ylabel("Frequency")
plt.title("Distribution of Total Pay")
plt.show()
```

```python
df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False).head(10) #Top Paying Job Titles
```

```python
df.groupby('Year')['TotalPay'].mean().plot(kind='line', marker='o') #Year salary trends
plt.title("Average Salary Trend by Year")
plt.ylabel("Average Total Pay")
plt.show()
```

```python
df[['BasePay','OvertimePay','OtherPay','TotalPay']].describe()
```

```python
plt.boxplot(df['TotalPay']) #outliers
plt.title("Total Pay")
plt.xlabel("Total Pay")
plt.show()
```

```python
cols = ['BasePay','OvertimePay','OtherPay','TotalPay']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce') # converts it to nums and missing values as nan
```

```python
df[cols].corr()
```

```python
df[['BasePay','OvertimePay','OtherPay','TotalPay']].corr()
sns.heatmap(df[['BasePay','OvertimePay','OtherPay','TotalPay']].corr(), annot=True, cmap="Blues")
plt.title("Correlation between Pay Components")
plt.show()
```

```python
top_jobs = df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False).head(10) #Top Paying Job Titles
sns.barplot(x=top_jobs.values, y=top_jobs.index, color="blue")
plt.title("Top 10 Highest Paying Job Titles")
plt.xlabel("Average Total Pay")
plt.ylabel("Job Title")
plt.show()
```

```python
top_agencies = df.groupby('Agency')['TotalPay'].mean().sort_values(ascending=False).head(10) #Top Paying Agencies
sns.barplot(x=top_agencies.values, y=top_agencies.index,color="grey")
plt.title("Top 10 Highest Paying Agencies")
plt.xlabel("Average Total Pay")
plt.ylabel("Agency")
plt.show()
```

```python
yearly_trend = df.groupby('Year')['TotalPay'].mean() #Salary Trends by Year
plt.figure(figsize=(8,5))
yearly_trend.plot(kind='line', marker='o')
plt.title("Average Total Pay by Year")
plt.xlabel("Year")
plt.ylabel("Average Total Pay")
plt.grid()
plt.show()
```

```python
components = df[['BasePay','OvertimePay','OtherPay','TotalPay']].mean() #Contribution of Pay Components
components[:-1].plot(kind='bar', color=['skyblue','orange','green'])
plt.title("Contribution of Pay Components (Average)")
plt.ylabel("Average Amount")
plt.show()
```
